```python
import os
import sys
import time
import numpy as np
import pandas as pd
import datetime as dtm
import matplotlib.pyplot as plt

try:
    root = os.path.dirname(os.path.abspath(__file__))
except:
    root = os.getcwd()
    
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 72
```

```python
old = pd.read_csv('radiance-forecast_original.csv', index_col=0)
old.index = pd.to_datetime(old.index)

new = pd.read_csv('radiance-forecast_new.csv', index_col=0)
new.index = pd.to_datetime(new.index)
```

```python
cols = sorted([c for c in new.columns if c.startswith('tsol')])
for c in cols:
    old[c].plot(color='C0')
    new[c].plot(color='C1', label='')
plt.legend()
plt.show()
```
